```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("label_data.csv")
df.head()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

scaler = StandardScaler()

# Fit and transform the data
standard_scaled_data = scaler.fit_transform(df)

# 다시 DataFrame으로 변환 (컬럼명 유지)
df_standard = pd.DataFrame(standard_scaled_data, columns=df.columns)

df_standard.head()
```

```python
scaler = MinMaxScaler()

# Fit and transform the data
minmax_scaled_data = scaler.fit_transform(df)

# 다시 DataFrame으로 변환 (컬럼명 유지)
df_minmax = pd.DataFrame(minmax_scaled_data, columns=df.columns)

df_minmax.head()
```

```python
scaler = RobustScaler()

# Fit and transform the data
robust_scaled_data = scaler.fit_transform(df)

# 다시 DataFrame으로 변환 (컬럼명 유지)
df_robust = pd.DataFrame(robust_scaled_data, columns=df.columns)

df_robust.head()
```
